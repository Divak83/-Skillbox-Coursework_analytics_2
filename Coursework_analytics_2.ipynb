### Предмет исследования №1:
### Daily Error Rate - Количество ошибок в день

```python
# Зависимость должности и количества ошибок 
pd.read_sql('SELECT ps."Position" AS position, COUNT(ps."Employee Name") AS ppl_count, SUM(ps."Daily Error Rate") AS error\
             FROM production_staff as ps\
             WHERE ps."Daily Error Rate" > 0\
             GROUP BY position ORDER BY error DESC', connect)
```

```python
# Зависимость количества ошибок от расы сотрудника
pd.read_sql('SELECT ps."Race Desc" AS race, COUNT(ps."Race Desc") AS ppl_count, SUM(ps."Daily Error Rate") AS count\
             FROM production_staff as ps\
             GROUP BY race ORDER BY count DESC', connect)
```

```python
# Зависимость кто у сотрудника менеджер и количество ошибок
pd.read_sql('SELECT ps."Manager Name" AS manager,\
             COUNT(ps."Employee Name") AS ppl_count,\
             SUM(ps."Daily Error Rate") AS err_count\
             FROM production_staff AS ps\
             WHERE ps."Daily Error Rate" > 0\
             GROUP BY manager ORDER BY err_count DESC', connect)
```

```python
# Зависимость зарплаты и кол-ва ошибок на должности 'Production Technician I'
pd.read_sql('SELECT ps."Employee Name" AS name, ps."Daily Error Rate" AS err_rate, ps."Pay" AS pay, sg."Hourly Mid" AS mid_pay\
             FROM production_staff AS ps\
             JOIN salary_grid AS sg\
             ON sg."Position" = ps."Position"\
             WHERE (sg."Position" = \'Production Technician I\' AND ps."Daily Error Rate" > 0)\
             ORDER BY err_rate DESC LIMIT 10',connect)
```

```python
# Зависимость зарплаты и кол-ва ошибок на должности 'Production Technician II'
pd.read_sql('SELECT ps."Employee Name" AS name, ps."Daily Error Rate" AS err_rate, ps."Pay" AS pay, sg."Hourly Mid" AS mid_pay\
             FROM production_staff AS ps\
             JOIN salary_grid AS sg\
             ON sg."Position" = ps."Position"\
             WHERE (sg."Position" = \'Production Technician II\' AND ps."Daily Error Rate" > 0)\
             ORDER BY err_rate DESC LIMIT 10',connect)
```

```python
# Зависимость пола и количества ошибок
pd.read_sql('SELECT hr."sex" AS sex, COUNT(hr."Employee Name") AS ppl_count, SUM(ps."Daily Error Rate") AS err_count\
             FROM hr_dataset AS hr\
             JOIN production_staff AS ps\
             ON ps."Employee Name" = hr."Employee Name"\
             WHERE ps."Daily Error Rate" > 0\
             GROUP BY sex',connect)
```

### Предмет исследования №2:
### Voluntarily Terminated - Увольнение по собственному желанию

```python
# Причины увольнений
pd.read_sql('SELECT hr."Reason For Term" AS reason, COUNT(hr."Employee Name") AS count\
             FROM hr_dataset AS hr\
             WHERE hr."Employment Status" = \'Voluntarily Terminated\'\
             GROUP BY reason ORDER BY count DESC', connect)
```

```python
# Зависимость увольнений от отдела 
pd.read_sql('SELECT hr."department" AS dept, COUNT(hr."Employee Name") AS count\
             FROM hr_dataset AS hr\
             WHERE hr."Employment Status" = \'Voluntarily Terminated\'\
             GROUP BY dept ORDER BY count DESC', connect)
```

```python
# Источнк найма и кол-во увольнений
pd.read_sql('SELECT hr."Employee Source" AS emp_sour, COUNT(hr."Employee Name") AS count\
             FROM hr_dataset AS hr\
             WHERE hr."Employment Status" = \'Voluntarily Terminated\'\
             GROUP BY emp_sour ORDER BY count DESC', connect)
```

```python
# Общее количество нанятых 
pd.read_sql('SELECT hr."Employee Source" AS emp_sour, COUNT(hr."Employee Name") AS count\
             FROM hr_dataset AS hr\
             GROUP BY emp_sour ORDER BY count DESC', connect)
```

```python
# Зависимость кто менеджер и кол-во увольнений
pd.read_sql('SELECT ps."Manager Name" AS manager, COUNT(ps."Employee Name") AS count\
             FROM production_staff AS ps\
             WHERE ps."Employment Status" = \'Voluntarily Terminated\'\
             GROUP BY manager ORDER BY count DESC', connect)
```

```python
# Общее кол-во сотрудников у менеджера
pd.read_sql('SELECT ps."Manager Name" AS manager, COUNT(ps."Employee Name") AS ppl_count\
             FROM production_staff AS ps\
             GROUP BY manager ORDER BY ppl_count DESC', connect)
```

```python
# Зависимость кол-ва увольнений от возраста
pd.read_sql('SELECT ps."Reason for Term" AS reason, MIN(hr."age") AS min_age, ROUND(AVG(hr."age")) AS avg_age,\
             MAX(hr."age") AS max_age\
             FROM production_staff AS ps\
             JOIN hr_dataset AS hr\
             ON ps."Employee Name" = hr."Employee Name"\
             WHERE ps."Employment Status" = \'Voluntarily Terminated\'\
             GROUP BY reason ORDER BY avg_age DESC',connect)
```

```python
# Зависимость семейного положения и количества увольнений
pd.read_sql('SELECT hr."maritaldesc" AS marital_status, COUNT(ps."Reason for Term") AS count\
             FROM hr_dataset AS hr\
             JOIN production_staff AS ps\
             ON ps."Employee Name" = hr."Employee Name"\
             WHERE ps."Employment Status" = \'Voluntarily Terminated\'\
             GROUP BY marital_status ORDER BY count DESC',connect)
```

### Предмет исследования №3:
### Days Employed - Количество отработаных дней 

```python
# Минимальная, средняя и максимальная продолжительность работы и возраст сотрудников в отделах
pd.read_sql('SELECT hr."department" AS dept, COUNT(hr."Employee Name") AS ppl_count,\
             MIN(hr."Days Employed") AS min_days, MAX(hr."Days Employed") AS max_days,\
             ROUND(AVG(hr."Days Employed")) AS avg_days, MIN(hr."age") AS min_age,\
             ROUND(AVG(hr."age")) AS avg_age, MAX(hr."age") AS max_age\
             FROM hr_dataset AS hr\
             GROUP BY dept ORDER BY avg_days DESC', connect)
```

```python
# Зависимость семейного положения и продолжительности работы
pd.read_sql('SELECT hr."maritaldesc" AS marital_status, MIN(hr."Days Employed") AS min_days,\
             MAX(hr."Days Employed") AS max_days,ROUND(AVG(hr."Days Employed")) AS avg_days\
             FROM hr_dataset AS hr\
             GROUP BY marital_status ORDER BY avg_days DESC', connect)
```

```python
# Зависимость кто менеджер и продолжительность работы сотрудника
pd.read_sql('SELECT hr."Manager Name" AS manager, MIN(hr."Days Employed") AS min_days,\
             MAX(hr."Days Employed") AS max_days, ROUND(AVG(hr."Days Employed")) AS avg_days\
             FROM hr_dataset AS hr\
             GROUP BY manager ORDER BY avg_days DESC', connect)
```

```python
# Распределение менеджеров по отделам
pd.read_sql('SELECT hr."Manager Name" AS manager, hr."department" AS dept\
             FROM hr_dataset AS hr\
             GROUP BY manager, dept ORDER BY manager', connect)
```

```python
# Зависимость продолжительности работы от пола сотрудника
pd.read_sql('SELECT hr."sex" AS sex, MIN(hr."Days Employed") AS min_days,\
             MAX(hr."Days Employed") AS max_days, ROUND(AVG(hr."Days Employed")) AS avg_days\
             FROM hr_dataset as hr\
             GROUP BY sex',connect)
```

```python
# Зависимость продолжительности работы от расы сотрудника
pd.read_sql('SELECT hr."racedesc" AS race, MIN(hr."Days Employed") AS min_days,\
             MAX(hr."Days Employed") AS max_days, ROUND(AVG(hr."Days Employed")) AS avg_days\
             FROM hr_dataset as hr\
             GROUP BY race ORDER BY avg_days DESC',connect)
```
